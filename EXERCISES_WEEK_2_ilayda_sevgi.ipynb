```python
import numpy as np
import matplotlib.pyplot as plt

# n değerlerini tanımla
n_values = np.linspace(1, 1000, 1000)  # n = 1, 2, ..., 1000

# Sine fonksiyonunun ve terimlerin hesaplanması
sin_n = np.sin(n_values)
sine_over_n = sin_n / n_values

# Alt ve üst sınırların hesaplanması
lower_bound = -1 / n_values  # Alt sınır: -1/n
upper_bound = 1 / n_values   # Üst sınır: 1/n

# Grafik oluşturulması
plt.figure(figsize=(10, 6))

# Verilen fonksiyon ve sınırların çizilmesi
plt.plot(n_values, sine_over_n, label=r'$\frac{\sin(n)}{n}$', color='blue')
plt.plot(n_values, lower_bound, label=r'$-\frac{1}{n}$', linestyle='--', color='red')
plt.plot(n_values, upper_bound, label=r'$\frac{1}{n}$', linestyle='--', color='green')

# Etiketler ve başlık
plt.xlabel('n')
plt.ylabel('Değer')
plt.title(r'Squeeze Teoremi: $\frac{\sin(n)}{n}$ ifadesinin $n \to \infty$ limitine yaklaşımı')
plt.legend()
plt.ylim(-0.2, 0.2)  # Y eksenini sınırlayarak daha iyi bir görünüm sağla
plt.grid(True)

# Grafiği göster
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate a_n = (1 + 1/n)^n
def a_n(n):
    return (1 + 1/n)**n

# Values of n from 1 to 1000
n_values = np.arange(1, 1001)

# Compute a_n for each n
a_n_values = a_n(n_values)

# Plot the values of a_n as n increases
plt.plot(n_values, a_n_values, label=r'$a_n = \left(1 + \frac{1}{n}\right)^n$', color='blue')
plt.axhline(y=np.exp(1), color='red', linestyle='--', label=r'$e \approx 2.718$')

# Add labels and title
plt.xlabel('n')
plt.ylabel('a_n')
plt.title('Convergence of the sequence $(1 + 1/n)^n$ to $e$')

# Show the legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()

# Optionally, print the value for large n (n = 1000)
print(f"a_n for n = 1000: {a_n(1000)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(x):
    return (x**2 - 1) / (x**2 + 1)

def g(x):
    return np.sin(x) / (x**2 + 1)

# Create an array of x values
x = np.linspace(-10, 10, 1000)

# Compute y values for both functions
y_f = f(x)
y_g = g(x)

# Plot f(x)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, y_f, label=r'$f(x) = \frac{x^2 - 1}{x^2 + 1}$')
plt.axhline(1, color='r', linestyle='--', label='y = 1 (Horizontal Asymptote)')
plt.title(r'Function $f(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# Plot g(x)
plt.subplot(1, 2, 2)
plt.plot(x, y_g, label=r'$g(x) = \frac{\sin(x)}{x^2 + 1}$')
plt.axhline(0, color='r', linestyle='--', label='y = 0 (Horizontal Asymptote)')
plt.title(r'Function $g(x)$')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
