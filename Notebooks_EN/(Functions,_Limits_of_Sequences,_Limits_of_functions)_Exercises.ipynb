\textbf{19. Limits of Sequences}
$$

\textbf{Task 1}  
Calculate:  
$$
\lim_{{n \to \infty}} \frac{n^2 + 3n}{2n^2 - 2n}
$$

Simplifying the expression:
$$
\frac{n^2 + 3n}{2n^2 - 2n} = \frac{n^2(1 + \frac{3}{n})}{n^2(2 - \frac{2}{n})}
$$
As \(n \to \infty\), the terms involving \( \frac{1}{n} \) approach zero, and the limit is:
$$
\lim_{{n \to \infty}} \frac{n^2(1 + \frac{3}{n})}{n^2(2 - \frac{2}{n})} = \frac{1}{2}
$$

---

\textbf{Task 2}  
Calculate:  
$$
\lim_{{n \to \infty}} \frac{(2n + 3)}{3n^3 - 1}
$$

As \(n \to \infty\), the highest degree terms dominate:
$$
\lim_{{n \to \infty}} \frac{2n + 3}{3n^3 - 1} = \lim_{{n \to \infty}} \frac{2n}{3n^3} = \lim_{{n \to \infty}} \frac{2}{3n^2} = 0
$$

\textbf{Task 3}  
Prove using the squeeze theorem:  
$$
\lim_{{n \to \infty}} \frac{\sin(n)}{n}
$$

We know that:
$$
-1 \leq \sin(n) \leq 1
$$
Thus:
$$
\frac{-1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
$$
As \( n \to \infty \), both \( \frac{-1}{n} \) and \( \frac{1}{n} \) approach 0. Therefore, by the squeeze theorem:
$$
\lim_{{n \to \infty}} \frac{\sin(n)}{n} = 0
$$

\textbf{20. Limits of Real Functions}
$$

\textbf{Task 1}  
Compute:  
$$
\lim_{{x \to \infty}} \frac{x^3 + 2x^2}{x^4 - 3x^3}
$$

Dividing both the numerator and denominator by \( x^4 \):
$$
\lim_{{x \to \infty}} \frac{x^3 + 2x^2}{x^4 - 3x^3} = \lim_{{x \to \infty}} \frac{\frac{x^3}{x^4} + \frac{2x^2}{x^4}}{1 - \frac{3x^3}{x^4}} = \lim_{{x \to \infty}} \frac{\frac{1}{x} + \frac{2}{x^2}}{1 - \frac{3}{x}} = 0
$$

\textbf{Task 2}  
Find:  
$$
\lim_{{x \to 0}} \frac{\sin(3x)}{2x + 1}
$$

Using the fact that \( \lim_{{x \to 0}} \sin(3x) = 3x \), we have:
$$
\lim_{{x \to 0}} \frac{\sin(3x)}{2x + 1} = \lim_{{x \to 0}} \frac{3x}{2x + 1} = \frac{0}{1} = 0
$$

\textbf{Task 3}  
Find the asymptotes of the function:  

1. For \( f(x) = \frac{x^2 - 1}{x^2 + 1} \), the horizontal asymptote is:
$$
\lim_{{x \to \infty}} f(x) = \lim_{{x \to \infty}} \frac{x^2 - 1}{x^2 + 1} = \frac{1 - 0}{1 + 0} = 1
$$
Thus, the horizontal asymptote is \( y = 1 \).

2. For \( g(x) = \frac{\sin(x)}{x^2 + 1} \), the horizontal asymptote is:
$$
\lim_{{x \to \infty}} g(x) = \lim_{{x \to \infty}} \frac{\sin(x)}{x^2 + 1} = 0
$$
Thus, the horizontal asymptote is \( y = 0 \).
