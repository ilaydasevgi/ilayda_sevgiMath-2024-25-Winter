# Canonical Equations of Conic Sections

### 1. **Circle**

The canonical equation of a **circle** with center \( (h, k) \) and radius \( r \) is:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

- **Example**: \$ (x - 2)^2 + (y + 3)^2 = 9 \$

### 2. **Ellipse**

The canonical equation of an **ellipse** with center \( (h, k) \), semi-major axis \( a \), and semi-minor axis \( b \) is:

$$
\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
$$

- **Example**: \$\frac{(x - 1)^2}{4} + \frac{(y - 2)^2}{9} = 1 \$

### 3. **Hyperbola**

The canonical equation of a **hyperbola** with center \( (h, k) \), transverse axis length \( 2a \), and conjugate axis length \( 2b \) is:

$$
\frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1
$$

- **Example**: \$ \frac{(x - 1)^2}{4} - \frac{(y - 2)^2}{9} = 1 \$

### 4. **Parabola**

The canonical equation of a **parabola** with vertex \( (h, k) \) and focal distance \( p \) is:

- Vertical parabola (opens up or down):

$$
(y - k) = a(x - h)^2
$$

- Horizontal parabola (opens left or right):

$$
(x - h) = a(y - k)^2
$$

- **Example**: \$ y = x^2 \$

### 5. **General Form of a Conic Section**

A general conic section equation is:

$$
Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0
$$

- **Circle**: \$ A = C \$, \$ B = 0 \$
- **Ellipse**: \$ A \neq C \$, \$ B = 0 \$
- **Hyperbola**: \$ B^2 - 4AC > 0 \$
- **Parabola**: $ B^2 - 4AC = 0 \$

### 6. **Foci and Eccentricity**

- **Circle**: \$ e = 0 \$
- **Ellipse**: \$ e = \sqrt{1 - \frac{b^2}{a^2}} \$
- **Hyperbola**: \$ e = \sqrt{1 + \frac{b^2}{a^2}} \$
- **Parabola**: \$ e = 1 \$

# Equations of Second-Order Curves with a Shifted Center

### 1. **Circle**

The canonical equation of a **circle** with a center at \( (h, k) \) and radius \( r \) is:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

- **Example**: \$ (x - 3)^2 + (y + 4)^2 = 16 $

### 2. **Ellipse**

The canonical equation of an **ellipse** with center \( (h, k) \), semi-major axis \( a \), and semi-minor axis \( b \) is:

$$
\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
$$

- **Example**: \$ \frac{(x - 2)^2}{9} + \frac{(y - 1)^2}{16} = 1 $

### 3. **Hyperbola**

The canonical equation of a **hyperbola** with center \( (h, k) \), transverse axis length \( 2a \), and conjugate axis length \( 2b \) is:

$$
\frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1
$$

- **Example**: \$ \frac{(x - 4)^2}{25} - \frac{(y - 5)^2}{16} = 1 \$

### 4. **Parabola**

The canonical equation of a **parabola** with vertex \( (h, k) \) and focal distance \( p \) is:

- Vertical parabola (opens up or down):

$$
y - k = a(x - h)^2
$$

- Horizontal parabola (opens left or right):

$$
x - h = a(y - k)^2
$$

- **Example**: \$ y - 2 = 3(x - 1)^2 \$

### 5. **General Form of a Conic Section**

A general second-order conic section equation with a shifted center \( (h, k) \) is:

$$
A(x - h)^2 + B(x - h)(y - k) + C(y - k)^2 + D(x - h) + E(y - k) + F = 0
$$

- This equation represents a circle, ellipse, hyperbola, or parabola depending on the values of \( A \), \( B \), and \( C \).

### 6. **Foci and Eccentricity for Shifted Conics**

- **Circle**: \$ e = 0 \$
- **Ellipse**: \$ e = \sqrt{1 - \frac{b^2}{a^2}} \$
- **Hyperbola**: \$ e = \sqrt{1 + \frac{b^2}{a^2}} \$
- **Parabola**: \$ e = 1 \$

# Reduction of the General Second-Order Curve Equation to Canonical Form

### 1. **General Form of Second-Order Curve**

The general second-order equation of a conic section is:

$$
Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0
$$

Where:
- \( A, B, C, D, E, F \) are constants.
- The type of curve (circle, ellipse, hyperbola, or parabola) depends on the values of \( A \), \( B \), and \( C \).

### 2. **Steps to Reduce to Canonical Form**

#### **Step 1: Eliminate the \( Bxy \) Term (Rotation)**

To eliminate the \( Bxy \) term, we rotate the coordinate axes by an angle \$ \theta \$. The transformation equations are:

$$
x' = x \cos \theta + y \sin \theta
$$
$$
y' = -x \sin \theta + y \cos \theta
$$

The angle \$ \theta \$ is chosen such that:

$$
\tan(2\theta) = \frac{B}{A - C}
$$

After this rotation, the equation becomes:

$$
A'x'^2 + C'y'^2 + D'x' + E'y' + F = 0
$$

Where \( A' \) and \( C' \) are the new coefficients, and \( Bxy \) is removed.

#### **Step 2: Translate the Center (Shift)**

If the curve is not centered at the origin, translate the origin to the center \( (h, k) \) by using the following substitutions:

$$
x = x' + h
$$
$$
y = y' + k
$$

Substitute these into the equation and simplify. The result is a shifted center at \( (h, k) \).

#### **Step 3: Complete the Square**

Now, we complete the square for the quadratic terms in \( x \) and \( y \) to get the canonical form.

For example, for a circle:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

For an ellipse, hyperbola, or parabola, the terms are adjusted accordingly.

### 3. **Canonical Forms After Reduction**

- **Circle**: \$ (x - h)^2 + (y - k)^2 = r^2 \$
- **Ellipse**: \$ \frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1 \$
- **Hyperbola**: \$ \frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1 \$
- **Parabola**: $ y - k = a(x - h)^2 \ or \   x - h = a(y - k)^2 \$

### 4. **Conditions for Different Conics**

- **Circle**: \$ A = C \$ and \$ B = 0 \$
- **Ellipse**: \$ A \neq C \$, \$ B = 0 \$, and \$ A, C > 0 \$
- **Hyperbola**: \$ B^2 - 4AC > 0 \$
- **Parabola**: \$ B^2 - 4AC = 0 \$

#### **1. Find the equation of a circle with center at point \( A(1, 2) \) and radius \( r = 3 \).**

The general equation of a circle is:
$$
(x - h)^2 + (y - k)^2 = r^2
$$
where \( (h, k) \) is the center of the circle and \( r \) is the radius.

- Center: \( (h, k) = (1, 2) \)
- Radius: \( r = 3 \)

Substituting into the formula:
$$
(x - 1)^2 + (y - 2)^2 = 3^2
$$

Simplify:
$$
(x - 1)^2 + (y - 2)^2 = 9
$$

---

#### **2. Find the equation of a parabola intersecting the \( Ox \)-axis at points \( x = 2, x = 4 \), and passing through point \( y(3) = 1 \).**

The general equation of a parabola with roots \( x = a \) and \( x = b \) is:
$$
y = k(x - a)(x - b)
$$

- Roots: \( x = 2, x = 4 \)
- Equation: \( y = k(x - 2)(x - 4) \)

Substitute the point \( (3, 1) \):
$$
1 = k(3 - 2)(3 - 4)
$$

Simplify:
$$
1 = k(1)(-1) \quad \Rightarrow \quad k = -1
$$

The equation becomes:
$$
y = -1(x - 2)(x - 4)
$$

Simplify further:
$$
y = -(x^2 - 6x + 8)
$$
$$
y = -x^2 + 6x - 8
$$

---

#### **3. Find the center of the ellipse with the equation \$ x^2 + 4y^2 - 4x - 16y + 16 = 0 \$.**

The general equation of an ellipse is:
$$
\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
$$
where \( (h, k) \) is the center.

1. Rewrite the equation by completing the square:
   $$
   x^2 - 4x + 4 + 4(y^2 - 4y + 4) = 16 + 4 + 16
   $$

2. Simplify:
   $$
   (x - 2)^2 + 4(y - 2)^2 = 36
   $$

3. Divide through by 36:
   $$
   \frac{(x - 2)^2}{36} + \frac{(y - 2)^2}{9} = 1
   $$

- The center is \( (h, k) = (2, 2) \).

#### **4. Find the slope $( m > 0 )$ of the line $ y = mx - 5 $ that is tangent to the circle $ x^2 + y^2 = 1 $.**

For a line $ y = mx + c $ to be tangent to a circle $ x^2 + y^2 = r^2 $, the distance from the center to the line equals the radius:
$$
\text{Distance} = \frac{|c|}{\sqrt{1 + m^2}} = r
$$

- \( c = -5 \), \( r = 1 \)

Substitute into the formula:
$$
\frac{5}{\sqrt{1 + m^2}} = 1
$$

Square both sides:
$$
\frac{25}{1 + m^2} = 1
$$

Solve for $ m^2 $:
$$
m^2 = 24 \quad \Rightarrow \quad m = \sqrt{24} = 2\sqrt{6}
$$

---

#### **5. Find the intersection points of the hyperbola $ x^2 - y^2 = 1 $ with the ellipse's line $ x^2 + 4y^2 = 6 $.**

1. Solve the system:
   - $ x^2 - y^2 = 1 $
   - $ x^2 + 4y^2 = 6 $

2. Subtract the first equation from the second:
   $$
   (x^2 + 4y^2) - (x^2 - y^2) = 6 - 1
   $$

Simplify:
$$
5y^2 = 5 \quad \Rightarrow \quad y^2 = 1 \quad \Rightarrow \quad y = \pm 1
$$

3. Substitute $ y = \pm 1 $ into $ x^2 - y^2 = 1 $:
   - For \$ y = 1 $:
     $$
     x^2 - 1 = 1 \quad \Rightarrow \quad x^2 = 2 \quad \Rightarrow \quad x = \pm\sqrt{2}
     $$
   - For \$ y = -1 \$:
     $$
     x^2 - (-1)^2 = 1 \quad \Rightarrow \quad x^2 = 2 \quad \Rightarrow \quad x = \pm\sqrt{2}
     $$

Intersection points:
$$
(\sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1)
$$

---

#### **6. For the given hyperbola \$ x^2 - y^2 = 1 \$, find the distance between its branches.**

The asymptotes of $ x^2 - y^2 = 1 $ are:
$$
y = x \quad \text{and} \quad y = -x
$$

The shortest distance between the branches is at \( x = 0 \):
- At $ x = 0 $, $ y = \pm 1 $.
- Distance:
  $$
  d = |1 - (-1)| = 2
  $$
