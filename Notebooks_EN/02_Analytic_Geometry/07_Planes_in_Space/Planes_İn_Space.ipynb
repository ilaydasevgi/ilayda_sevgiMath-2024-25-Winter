# Planes and Lines in Space

## 1. **Planes in Space**

A **plane** in 3D space is a flat, two-dimensional surface that extends infinitely. It can be defined by a point and a normal vector perpendicular to the surface.

### **Equation of a Plane**

The general equation of a plane is given by:
$$ Ax + By + Cz = D $$

Where:
- $(A, B, C)$ is the normal vector to the plane,
- $(x, y, z)$ are the coordinates of any point on the plane,
- $D$ is a constant that determines the position of the plane.

### **Vector Equation of a Plane**

A plane can also be represented using vectors. The vector equation of a plane is:
$$ \mathbf{r} = \mathbf{r_0} + \lambda \mathbf{v_1} + \mu \mathbf{v_2} $$

Where:
- $\mathbf{r}$ is the position vector of any point on the plane,
- $\mathbf{r_0}$ is the position vector of a known point on the plane,
- $\mathbf{v_1}$ and $\mathbf{v_2}$ are two non-parallel vectors lying on the plane,
- $\lambda$ and $\mu$ are scalar parameters.

### **Plane Equation Passing Through Three Points**

Given three points $\mathbf{P_1}(x_1, y_1, z_1)$, $\mathbf{P_2}(x_2, y_2, z_2)$, and $\mathbf{P_3}(x_3, y_3, z_3)$ on the plane, we first calculate the vectors:
$$ \mathbf{v_1} = \mathbf{P_2} - \mathbf{P_1} $$
$$ \mathbf{v_2} = \mathbf{P_3} - \mathbf{P_1} $$

Then the normal vector $\mathbf{n}$ is given by the cross product:
$$ \mathbf{n} = \mathbf{v_1} \times \mathbf{v_2} $$

The equation of the plane is then:
$$ \mathbf{n} \cdot (\mathbf{r} - \mathbf{P_1}) = 0 $$

### **Plane Equation in Intercept Form**

The intercept form of the plane equation is:
$$ \frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1 $$

Where $a$, $b$, and $c$ are the intercepts on the x, y, and z axes, respectively.

### **Special Cases of Plane Positioning**

- **Parallel Planes**: Two planes are parallel if their normal vectors are proportional:
  $$ \mathbf{n_1} \parallel \mathbf{n_2} $$

- **Perpendicular Planes**: Two planes are perpendicular if their normal vectors are orthogonal:
  $$ \mathbf{n_1} \cdot \mathbf{n_2} = 0 $$

## 2. **Angle Between Two Planes**

The angle $\theta$ between two planes is determined by the angle between their normal vectors:
$$ \cos(\theta) = \frac{\mathbf{n_1} \cdot \mathbf{n_2}}{|\mathbf{n_1}| |\mathbf{n_2}|} $$

Where:
- $\mathbf{n_1}$ and $\mathbf{n_2}$ are the normal vectors of the two planes.

## 3. **Conditions for Parallelism and Perpendicularity of Two Planes**

- **Parallelism**: Two planes are parallel if their normal vectors are proportional:
  $$ \mathbf{n_1} \parallel \mathbf{n_2} $$

- **Perpendicularity**: Two planes are perpendicular if their normal vectors are orthogonal:
  $$ \mathbf{n_1} \cdot \mathbf{n_2} = 0 $$

## 4. **Line in Space**

A line in space can be represented by a parametric equation:
$$ \mathbf{r}(t) = \mathbf{r_0} + t\mathbf{v} $$

Where:
- $\mathbf{r_0}$ is a point on the line,
- $\mathbf{v}$ is the direction vector of the line,
- $t$ is a scalar parameter.

## 5. **Types of Line Equations in Space**

- **Parametric Form**:
  $$ \mathbf{r}(t) = \mathbf{r_0} + t\mathbf{v} $$

- **Symmetric Form**:
  $$ \frac{x - x_1}{a} = \frac{y - y_1}{b} = \frac{z - z_1}{c} $$

Where $(x_1, y_1, z_1)$ is a point on the line and $(a, b, c)$ is the direction vector.

## 6. **Relative Positions of a Line and a Plane**

The relative positions of a line and a plane can be:
1. **Intersecting**: The line intersects the plane at a single point.
2. **Parallel**: The line is parallel to the plane and does not intersect it.
3. **Coincident**: The line lies entirely on the plane.

## 7. **Relative Positions of Two Lines**

Two lines in space can be:
1. **Intersecting**: The lines intersect at a single point.
2. **Parallel**: The lines do not intersect and have the same direction.
3. **Skew**: The lines do not intersect and are not parallel (they lie in different planes).

## 8. **Angle Between a Line and a Plane**

The angle $\theta$ between a line and a plane is given by the complement of the angle between the line and the normal to the plane:
$$ \sin(\theta) = \frac{|\mathbf{v} \cdot \mathbf{n}|}{|\mathbf{v}| |\mathbf{n}|} $$

Where:
- $\mathbf{v}$ is the direction vector of the line,
- $\mathbf{n}$ is the normal vector of the plane.

### **1. The plane passes through points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$. Find the equation of the plane.**

#### Step 1: Find two vectors on the plane.
The vectors $\vec{AB}$ and $\vec{AC}$ are calculated as:
$ \vec{AB} = B - A = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2) $
$ \vec{AC} = C - A = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1) $

#### Step 2: Find the normal vector using the cross product.
The cross product $\vec{AB} \times \vec{AC}$ gives the normal vector to the plane:
$$
\vec{AB} \times \vec{AC} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
2 & 2 & 2 \\
1 & -1 & 1
\end{vmatrix} = \mathbf{i}(2 - (-2)) - \mathbf{j}(2 - 2) + \mathbf{k}(-2 - 2)
$$
$$
\vec{n} = (4, 0, -4)
$$

#### Step 3: Write the equation of the plane.
The equation of a plane is:
$ n_1(x - x_0) + n_2(y - y_0) + n_3(z - z_0) = 0 $
Substitute $\vec{n} = (4, 0, -4)$ and $A(1, 2, 3)$:
$ 4(x - 1) + 0(y - 2) - 4(z - 3) = 0 $
Simplify:
$ 4x - 4 - 4z + 12 = 0 \quad \implies \quad 4x - 4z + 8 = 0 $

**Answer:**
$ 4x - 4z + 8 = 0 $

### **2. The plane passes through point $A(1, 2, 3)$ and is parallel to the plane $2x + 3y + 4z = 5$.**

#### Step 1: Understand the problem.
Planes that are parallel share the same normal vector. The given plane $2x + 3y + 4z = 5$ has a normal vector $\vec{n} = (2, 3, 4)$.

#### Step 2: Write the equation using the point.
Using the formula for the plane:
$ n_1(x - x_0) + n_2(y - y_0) + n_3(z - z_0) = 0 $
Substitute $\vec{n} = (2, 3, 4)$ and $A(1, 2, 3)$:
$ 2(x - 1) + 3(y - 2) + 4(z - 3) = 0 $
Simplify:
$ 2x - 2 + 3y - 6 + 4z - 12 = 0 \quad \implies \quad 2x + 3y + 4z - 20 = 0 $

**Answer:**
$ 2x + 3y + 4z - 20 = 0 $

### **3. The plane passes through point $A(1, 2, 3)$ and is perpendicular to the normal vector $\vec{n} = [2, 3, 4]$.**

This question is solved in the same way as Question 2 since the plane is perpendicular to the same normal vector.

**Answer:**
$ 2x + 3y + 4z - 20 = 0 $

### **4. The line of intersection of the planes $2x + 3y + 4z = 5$ and $3x + 4y + 2z = 6$.**

#### Step 1: Find the direction vector of the line.
The direction vector of the line is perpendicular to the normal vectors of the two planes. Let:
$ \vec{n_1} = (2, 3, 4), \quad \vec{n_2} = (3, 4, 2) $
The direction vector is given by the cross product:
$$
\vec{n_1} \times \vec{n_2} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
2 & 3 & 4 \\
3 & 4 & 2
\end{vmatrix} = (-10, 8, -1)
$$

#### Step 2: Find a point on the line.
Substitute $z = 0$ into the equations:
$ 2x + 3y = 5, \quad 3x + 4y = 6 $
Solve:
$ x = -2, \quad y = 3 $
So, a point on the line is $(-2, 3, 0)$.

#### Step 3: Write the parametric equations.
The parametric equations of the line are:
$ x = -2 - 10t, \quad y = 3 + 8t, \quad z = -t $

**Answer:**
$ x = -2 - 10t, \quad y = 3 + 8t, \quad z = -t $

### **5. The plane passes through $A(1, 2, 3)$ and is parallel to $\vec{v_1} = [1, 0, 1]$ and $\vec{v_2} = [0, 1, -1]$.**

#### Step 1: Find the normal vector.
The normal vector is perpendicular to $\vec{v_1}$ and $\vec{v_2}$. Compute the cross product:
$$
\vec{v_1} \times \vec{v_2} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 0 & 1 \\
0 & 1 & -1
\end{vmatrix} = (-2, 1, 1)
$$

#### Step 2: Write the equation of the plane.
Using the normal vector $(-2, 1, 1)$ and point $A(1, 2, 3)$:
$ -2(x - 1) + 1(y - 2) + 1(z - 3) = 0 $
Simplify:
$ -2x + y + z - 3 = 0 $

**Answer:**
$ -2x + y + z - 3 = 0 $

### **6. Parallel and perpendicular planes to $2x + 3y + 4z = 5$.**

#### (a) Parallel Plane
Parallel planes share the same normal vector. Change the constant $c$ in the equation:
$ 2x + 3y + 4z = 7 $

#### (b) Perpendicular Plane
A perpendicular plane requires its normal vector to be perpendicular to $\vec{n} = (2, 3, 4)$. Let $\vec{n'} = (a, b, c)$:
$ 2a + 3b + 4c = 0 $
Choose $a = 1, b = 1$. Solve for $c$:
$ 4c = -5 \quad \implies \quad c = -\frac{5}{4} $
The perpendicular plane is:
$ x + y - \frac{5}{4}z = 0 $

**Answers:**
- Parallel plane: $ 2x + 3y + 4z = 7 $
- Perpendicular plane: $ x + y - \frac{5}{4}z = 0 $

### **7. Distance from point $A(1, 2, 3)$ to the plane $2x + 3y + 4z = 5$.**

#### Step 1: Use the distance formula.
The distance $d$ from a point $(x_1, y_1, z_1)$ to a plane $Ax + By + Cz + D = 0$ is:
$ d = \frac{|Ax_1 + By_1 + Cz_1 + D|}{\sqrt{A^2 + B^2 + C^2}} $

#### Step 2: Substitute values.
Substitute $A = 2, B = 3, C = 4, D = -5$ and $(x_1, y_1, z_1) = (1, 2, 3)$:
$ d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}} = \frac{|15|}{\sqrt{29}} $

**Answer:**
$ d = \frac{15}{\sqrt{29}} $

### **8. The plane intersects the axes at $A(2, 0, 0)$, $B(0, 3, 0)$, $C(0, 0, 4)$.**

#### Step 1: Write the general form of the plane.
A plane that intersects the axes at $(a, 0, 0)$, $(0, b, 0)$, $(0, 0, c)$ is:
$ \frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1 $

#### Step 2: Substitute the intercepts.
Substitute $a = 2, b = 3, c = 4$:
$ \frac{x}{2} + \frac{y}{3} + \frac{z}{4} = 1 $

**Answer:**
$ \frac{x}{2} + \frac{y}{3} + \frac{z}{4} = 1 $

### **9. Angle between the planes $x + y + z = 1$ and $x = 0$.**

#### Step 1: Use the formula for the angle between two planes.
The angle $\theta$ is given by:
$ \cos\theta = \frac{\vec{n_1} \cdot \vec{n_2}}{\|\vec{n_1}\| \|\vec{n_2}\|} $

#### Step 2: Identify normal vectors.
The normal vector of $x + y + z = 1$ is $\vec{n_1} = (1, 1, 1)$, and for $x = 0$, $\vec{n_2} = (1, 0, 0)$.

#### Step 3: Substitute into the formula.
$ \vec{n_1} \cdot \vec{n_2} = 1, \quad \|\vec{n_1}\| = \sqrt{3}, \quad \|\vec{n_2}\| = 1 $
$ \cos\theta = \frac{1}{\sqrt{3}} $

**Answer:**
$ \theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right) $

### **10. Vector perpendicular to the plane $x + y + z = 1$.**

The normal vector to the plane is:
$ \vec{n} = (1, 1, 1) $

**Answer:**
$ \vec{n} = (1, 1, 1) $
