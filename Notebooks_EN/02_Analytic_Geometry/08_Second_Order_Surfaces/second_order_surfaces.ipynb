## 1. General Equations of 3D Surfaces

### 1.1 Ellipsoid
The general equation of an ellipsoid centered at the origin is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1
$$
where:
- $a$, $b$, and $c$ are the semi-axes of the ellipsoid along the $x$, $y$, and $z$ axes, respectively.

#### Special Case: Sphere
When $a = b = c = r$, the ellipsoid becomes a sphere:
$$
x^2 + y^2 + z^2 = r^2
$$
where $r$ is the radius of the sphere.

---

### 1.2 Hyperboloid
#### One-Sheeted Hyperboloid
The general equation of a hyperboloid of one sheet is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1
$$
where:
- $a$, $b$, and $c$ are constants determining the shape and orientation of the hyperboloid.

#### Two-Sheeted Hyperboloid
The general equation of a hyperboloid of two sheets is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = -1
$$

For hyperboloids, the sign of the constant on the right-hand side determines whether the surface is one-sheeted or two-sheeted.

---

### 1.3 Sphere
A sphere is a special case of the ellipsoid where all semi-axes are equal:
$$
x^2 + y^2 + z^2 = r^2
$$
where $r$ is the radius of the sphere.

For a sphere with its center shifted to $(h, k, l)$:
$$
(x - h)^2 + (y - k)^2 + (z - l)^2 = r^2
$$

### **Task 1: Equation of a Sphere**

#### Problem:
Write the equation of a sphere with center at $ P = (1, 2, 3) $  and radius $  r = 3 $ .

#### Formula:
The general equation of a sphere is:
$$
(x - h)^2 + (y - k)^2 + (z - l)^2 = r^2
$$
where:
- $  (h, k, l) $  is the center of the sphere,
- $  r $  is the radius.

#### Solution:
Here, the center $  (h, k, l) = (1, 2, 3) $ , and the radius $  r = 3 $ . Substituting these values:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 3^2
$$

Simplify:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

#### Final Answer:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

### **Task 2: Do the Spheres Intersect?**

#### Problem:
Check if the spheres:
$$
x^2 + y^2 + z^2 = 1
$$
and
$$
x^2 + y^2 + z^2 = 2
$$
have any common points.

#### Observation:
Both equations represent spheres centered at \( (0, 0, 0) \):
1. The first sphere has radius $  r_1 = \sqrt{1} = 1 $ ,
2. The second sphere has radius $  r_2 = \sqrt{2} $ .
3. The spheres have the same center at $ C(0, 0, 0) $

#### Solution:
For the spheres to intersect, their radii must overlap. Here, the spheres are concentric (same center), but their radii differ:
- $  r_1 = 1 $ ,
- $  r_2 = \sqrt{2} \approx 1.41 $ .

Since $  r_1 < r_2 $ , and both spheres share the same center, the smaller sphere is entirely inside the larger one without touching its boundary.

#### Final Answer:
The spheres do not intersect and do not have any common points excpet their center point.

### **Task 3: Curve of Intersection Between Two Spheres**

#### Problem:
Find the curve formed by the intersection of the spheres:
$$
x^2 + y^2 + z^2 = 1
$$
and
$$
(x - 1)^2 + y^2 + z^2 = 1
$$

#### Step 1: Simplify the Second Sphere
Expand the second sphere equation:
$$
(x - 1)^2 + y^2 + z^2 = 1
$$
$$
x^2 - 2x + 1 + y^2 + z^2 = 1
$$
$$
x^2 + y^2 + z^2 - 2x = 0
$$

#### Step 2: Substitute the First Sphere Equation
The first sphere equation gives:
$$
x^2 + y^2 + z^2 = 1
$$

Substitute this into the simplified second sphere equation:
$$
1 - 2x = 0
$$

Solve for \( x \):
$$
x = \frac{1}{2}
$$

#### Step 3: Equation of the Curve
The curve of intersection lies on the plane $  x = \frac{1}{2} $  and also satisfies the first sphere equation:
$$
\left(\frac{1}{2}\right)^2 + y^2 + z^2 = 1
$$

Simplify:
$$
\frac{1}{4} + y^2 + z^2 = 1
$$
$$
y^2 + z^2 = \frac{3}{4}
$$

This is the equation of a circle in the plane $  x = \frac{1}{2} $ , centered at $  \left(\frac{1}{2}, 0, 0\right) $  with radius $  \sqrt{\frac{3}{4}} = \frac{\sqrt{3}}{2} $ .

#### Final Answer:
The curve is a circle:
$$
x = \frac{1}{2}, \quad y^2 + z^2 = \frac{3}{4}
$$

### **Task 4: Tangent Plane to a Paraboloid**

#### Problem:
Find the tangent plane to the paraboloid:
$$
z = (x - 1)^2 + y^2 + 1
$$
at the point \( P = (1, 0, 1) \).

#### Formula:
The equation of a tangent plane to a surface $  z = f(x, y) $  at a point $  (x_0, y_0, z_0) $  is:
$$
z - z_0 = f_x(x_0, y_0)(x - x_0) + f_y(x_0, y_0)(y - y_0)
$$
where:
- $  f_x = \frac{\partial f}{\partial x} $ ,
- $  f_y = \frac{\partial f}{\partial y} $ .

#### Step 1: Compute Partial Derivatives
The paraboloid is:
$$
z = f(x, y) = (x - 1)^2 + y^2 + 1
$$

Compute $  f_x $ :
$$
f_x = \frac{\partial}{\partial x}\left[(x - 1)^2 + y^2 + 1\right] = 2(x - 1)
$$

Compute $  f_y $ :
$$
f_y = \frac{\partial}{\partial y}\left[(x - 1)^2 + y^2 + 1\right] = 2y
$$

#### Step 2: Evaluate at \( P = (1, 0, 1) \)
At \( P = (1, 0, 1) \):
- $  f_x(1, 0) = 2(1 - 1) = 0 $ ,
- $  f_y(1, 0) = 2(0) = 0 $ .

#### Step 3: Tangent Plane Equation
The tangent plane equation becomes:
$$
z - z_0 = f_x(x_0, y_0)(x - x_0) + f_y(x_0, y_0)(y - y_0)
$$

Substitute:
$$
z - 1 = 0(x - 1) + 0(y - 0)
$$

Simplify:
$$
z = 1
$$

#### Final Answer:
The equation of the tangent plane is:
$$
z = 1
$$
