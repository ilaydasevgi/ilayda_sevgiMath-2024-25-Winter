# Notes: Useful Formulas for Extremum Problems

## 1. Critical Points and Extremum
To find the extremum (maximum or minimum) of a function $f(x)$:
1. Find the derivative: $f'(x)$.
2. Set $f'(x) = 0$ to find the critical points.
3. Use the second derivative test:
   - If $f''(x) > 0$ at a critical point, it is a local minimum.
   - If $f''(x) < 0$ at a critical point, it is a local maximum.
   - If $f''(x) = 0$, the test is inconclusive.

## 2. Vertex Formula for Quadratic Functions
For a quadratic function $f(x) = ax^2 + bx + c$, the $x$-coordinate of the vertex (extremum point) is:
$$
x = -\frac{b}{2a}
$$
- If $a > 0$, the vertex is a minimum.
- If $a < 0$, the vertex is a maximum.

## 3. Optimization with Constraints
For a rectangle with a fixed perimeter $P$:
- Let the dimensions of the rectangle be $x$ and $y$.
- The perimeter constraint is $2x + 2y = P$.
- The area of the rectangle is $A = x \cdot y$.
- Substitute $y = \frac{P}{2} - x$ into the area formula:
$$
A(x) = x \left(\frac{P}{2} - x\right)
$$
Then, find the maximum area by finding the critical points of $A(x)$.

## 4. Rational Functions
For a function $f(x) = \frac{g(x)}{h(x)}$:
1. Differentiate using the quotient rule:
$$
f'(x) = \frac{g'(x)h(x) - g(x)h'(x)}{[h(x)]^2}
$$
2. Set $f'(x) = 0$ and solve for $x$ to find critical points.
3. Check the behavior around these points to determine maxima or minima.

# Solutions to Extremum Problems

---

## Problem 6

### Given:
The profit function:
$$P(u) = -2u^2 + 50u - 300$$

### Steps:
1. **Find the derivative**:
   Differentiate $P(u)$:
   $$P'(u) = -4u + 50$$

2. **Find critical points**:
   Set $P'(u) = 0$:
   $$-4u + 50 = 0$$
   $$u = \frac{50}{4} = 12.5$$

3. **Verify maximum**:
   Since the coefficient of $u^2$ in $P(u)$ is negative $(-2)$, the parabola opens downward. Therefore, the critical point at $u = 12.5$ is a **maximum**.

### Solution:
The number of units that maximizes profit is:
$$u = 12.5 \, \text{units}$$

## Problem 7:

### Given:
The total string length (perimeter) is:
$$2x + 2y = 10 \implies y = 5 - x$$

### Steps:
1. **Write the area function**:
   The area of the rectangle is:
   $$A(x) = x \cdot y = x \cdot (5 - x) = 5x - x^2$$

2. **Differentiate the area function**:
   $$A'(x) = 5 - 2x$$

3. **Find critical points**:
   Set $A'(x) = 0$:
   $$5 - 2x = 0$$
   $$x = \frac{5}{2} = 2.5$$

4. **Find the corresponding $y$**:
   Substitute $x = 2.5$ into $y = 5 - x$:
   $$y = 5 - 2.5 = 2.5$$

5. **Verify maximum area**:
   The function $A(x) = 5x - x^2$ is a parabola with a negative leading coefficient $(-x^2)$. Therefore, it has a maximum at $x = 2.5$.

### Solution:
The dimensions of the rectangle are:
$$x = 2.5 \, \text{m}, \quad y = 2.5 \, \text{m}$$

## Problem 8: Extremum of $f(x) = x^2 + 3x - 5$

### Given:
The quadratic function:
$$f(x) = x^2 + 3x - 5$$

### Steps:
1. **Find the derivative**:
   Differentiate $f(x)$:
   $$f'(x) = 2x + 3$$

2. **Find critical points**:
   Set $f'(x) = 0$:
   $$2x + 3 = 0$$
   $$x = -\frac{3}{2}$$

3. **Second derivative test**:
   Find the second derivative:
   $$f''(x) = 2$$
   Since $f''(x) > 0$, the function is concave upward, and $x = -\frac{3}{2}$ is a **minimum**.

### Solution:
The function has a **minimum** at:
$$x = -\frac{3}{2}$$

## Problem 9: Extremum of $f(x) = \frac{x^2 + 2x + 1}{x - 1}$

### Given:
The rational function:
$$f(x) = \frac{x^2 + 2x + 1}{x - 1}$$

### Steps:
1. **Find the derivative**:
   Use the quotient rule:
   $$f'(x) = \frac{(2x + 2)(x - 1) - (x^2 + 2x + 1)(1)}{(x - 1)^2}$$

2. **Simplify the numerator**:
   Expand the numerator:
   $$f'(x) = \frac{2x^2 - 2x + 2x - 2 - (x^2 + 2x + 1)}{(x - 1)^2}$$
   $$f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2}$$

3. **Find critical points**:
   Set $f'(x) = 0$, which means solving the numerator:
   $$x^2 - 2x - 3 = 0$$
   Factorize:
   $$(x - 3)(x + 1) = 0$$
   $$x = 3, \, x = -1$$

4. **Verify second derivative**:
   Further analysis using the second derivative can classify these points, but it is clear that:
   - $x = 3$ and $x = -1$ are critical points.
   - Both are in the domain of the function $(x \neq 1)$.

### Solution:
The critical points are:
$$x = 3, \, x = -1$$
