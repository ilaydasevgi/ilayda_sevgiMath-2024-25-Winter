# **Integrals: Academic Notes**

## **Definition of Integrals**
1. **Indefinite Integral:**
   The indefinite integral of a function $f(x)$ is the antiderivative of $f(x)$, represented as:
   $$
   \int f(x) \, dx = F(x) + C,
   $$
   where $F(x)$ is the antiderivative of $f(x)$ and $C$ is the constant of integration.

2. **Definite Integral:**
   The definite integral of a function $f(x)$ over an interval $[a, b]$ is defined as:
   $$
   \int_a^b f(x) \, dx = F(b) - F(a),
   $$
   where $F(x)$ is the antiderivative of $f(x)$.

---

## **Fundamental Theorem of Calculus**
1. **First Part:**
   If $f(x)$ is continuous on $[a, b]$ and $F(x)$ is its antiderivative, then:
   $$
   \int_a^b f(x) \, dx = F(b) - F(a).
   $$

2. **Second Part:**
   If $F(x)$ is an antiderivative of $f(x)$, then:
   $$
   \frac{d}{dx} \int_a^x f(t) \, dt = f(x).
   $$

---

## **Basic Integration Rules**
1. **Power Rule:**
   $$
   \int x^n \, dx = \frac{x^{n+1}}{n+1} + C, \quad \text{for } n \neq -1.
   $$

2. **Constant Rule:**
   $$
   \int c \, dx = cx + C,
   $$
   where \(c\) is a constant.

3. **Sum/Difference Rule:**
   $$
   \int [f(x) \pm g(x)] \, dx = \int f(x) \, dx \pm \int g(x) \, dx.
   $$

4. **Exponential Function:**
   $$
   \int e^x \, dx = e^x + C.
   $$

5. **Trigonometric Functions:**
   - $$
     \int \sin(x) \, dx = -\cos(x) + C.
     $$

   - $$
     \int \cos(x) \, dx = \sin(x) + C.
     $$

   - $$
     \int \sec^2(x) \, dx = \tan(x) + C.
     $$

   - $$
     \int \csc^2(x) \, dx = -\cot(x) + C.
     $$

6. **Logarithmic Function:**
   $$
   \int \frac{1}{x} \, dx = \ln|x| + C.
   $$

---

## **Special Integration Techniques**
1. **Substitution Rule:**
   If $u = g(x)$, then:
   $$
   \int f(g(x))g'(x) \, dx = \int f(u) \, du.
   $$

2. **Integration by Parts:**
   For functions $u(x)$ and $v(x)$:
   $$
   \int u \, dv = uv - \int v \, du.
   $$

3. **Trigonometric Identities:**
   Use these to simplify integrals involving trigonometric functions:
   - $\sin^2(x) + \cos^2(x) = 1$,
   - $\tan^2(x) + 1 = \sec^2(x)$,
   - $\sec^2(x) - 1 = \tan^2(x)$.

4. **Partial Fraction Decomposition:**
   If $f(x)$ is a rational function, decompose it into simpler fractions before integrating.

---

## **Applications of Definite Integrals**
1. **Area Under a Curve:**
   The area under $f(x)$ over $[a, b]$ is:
   $$
   A = \int_a^b f(x) \, dx.
   $$

2. **Volume of a Solid of Revolution:**
   - Using the disk method:
     $$
     V = \pi \int_a^b [f(x)]^2 \, dx.
     $$
   - Using the shell method:
     $$
     V = 2\pi \int_a^b x \cdot f(x) \, dx.
     $$

3. **Arc Length of a Curve:**
   The arc length of $y = f(x)$ over $[a, b]$ is:
   $$
   L = \int_a^b \sqrt{1 + [f'(x)]^2} \, dx.
   $$

4. **Area Between Two Curves:**
   The area between $f(x)$ and $g(x)$ over $[a, b]$ is:
   $$
   A = \int_a^b [f(x) - g(x)] \, dx, \quad \text{where } f(x) \geq g(x).
   $$

---

## **Special Cases Related to the Questions**

### **1. Integral of a Sum/Difference**
For example:
$$
\int (5 \sin(x) + 3 e^x) \, dx = 5 \int \sin(x) \, dx + 3 \int e^x \, dx.
$$

### **2. Trigonometric Simplification**
For $\int (\sin^2(x) + \cos^2(x)) \, dx$:
- Use $\sin^2(x) + \cos^2(x) = 1$:
$$
\int (\sin^2(x) + \cos^2(x)) \, dx = \int 1 \, dx = x + C.
$$

### **3. Substitution Example**
For $\int \frac{\cos(\ln(x))}{x} \, dx$:
- Let $u = \ln(x)$, then $du = \frac{1}{x} dx$:
$$
\int \frac{\cos(\ln(x))}{x} \, dx = \int \cos(u) \, du = \sin(u) + C = \sin(\ln(x)) + C.
$$

## **1(a): $ \int 1 dx $**

## **1(b): $ \int (x^2 + 2) dx $**

1. **Split the integral**:
   $$ \int (x^2 + 2) dx = \int x^2 dx + \int 2 dx $$

2. **Solve each part**:
   - For $ \int x^2 dx $:

     $$ \int x^n dx = \frac{x^{n+1}}{n+1} + C $$
     Here $n = 2$, so:

     $$ \int x^2 dx = \frac{x^3}{3} $$
   - For $ \int 2 dx $:

     $$ \int a dx = ax + C $$
     Here $a = 2$, so:

     $$ \int 2 dx = 2x $$

3. **Combine results**:
   $$ \int (x^2 + 2) dx = \frac{x^3}{3} + 2x + C $$

3. **Final result**:

3. **Final result**:

## **1(m): $ \int x \ln(x) dx $**

### **6th question**
Find the distance of the moving particle between time $ t = 0 $ and $ t = 2 $ for the following position function:

$$
x(t) = 3t^2 - 6t + 1
$$

---

### **Step-by-Step Solution:**

We are given the position function of a particle $ x(t) = 3t^2 - 6t + 1 $, and we are tasked with finding the total distance traveled by the particle between $ t = 0 $ and $ t = 2 $.

The distance traveled by the particle can be computed by integrating the **absolute value** of the velocity function over the given time interval. The reason we use the absolute value is because distance is always positive, even if the particle changes direction.

The steps are as follows:

---

### **Step 1: Find the velocity function $ v(t) $**

The velocity is the derivative of the position function $ x(t) $:

$$
v(t) = \frac{dx}{dt}
$$

Differentiating $ x(t) = 3t^2 - 6t + 1 $:

$$
v(t) = 6t - 6
$$

---

### **Step 2: Determine if the velocity changes sign between $ t = 0 $ and $ t = 2 $**

Since we need to take the absolute value of the velocity, let's check if the velocity function changes sign within the given time interval.

Set the velocity function equal to zero to find the time $ t $ at which the particle changes direction:

$$
6t - 6 = 0
$$

Solving for $ t $:

$$
6t = 6
$$

$$
t = 1
$$

Thus, the velocity changes sign at $ t = 1 $.

---

### **Step 3: Split the integral at $ t = 1 $**

To calculate the total distance, we need to break the integral into two parts: one from $ t = 0 $ to $ t = 1 $ and the other from $ t = 1 $ to $ t = 2 $. This is because the velocity changes sign at $ t = 1 $, so we need to account for the absolute value of the velocity.

For $ t \in [0, 1] $, the velocity is negative, so the distance traveled is the integral of $ -v(t) $ over this interval. For $ t \in [1, 2] $, the velocity is positive, so the distance traveled is simply the integral of $ v(t) $ over this interval.

Thus, the total distance $ D $ is:

$$
D = \int_0^1 |v(t)| \, dt + \int_1^2 |v(t)| \, dt
$$

Since $ v(t) $ is negative in the first interval, we use $ -v(t) $ for the integral from $ 0 $ to $ 1 $, and we use $ v(t) $ for the integral from $ 1 $ to $ 2 $.

---

### **Step 4: Compute the integrals**

#### **Interval $ [0, 1] $:**

We have:

$$
D_1 = \int_0^1 -v(t) \, dt = \int_0^1 -(6t - 6) \, dt = \int_0^1 (-6t + 6) \, dt
$$

To solve this integral:

$$
D_1 = \left[ -3t^2 + 6t \right]_0^1
$$

Evaluating at the bounds:

$$
D_1 = \left[ -3(1)^2 + 6(1) \right] - \left[ -3(0)^2 + 6(0) \right]
$$

$$
D_1 = (-3 + 6) - (0) = 3
$$

#### **Interval $ [1, 2] $:**

We have:

$$
D_2 = \int_1^2 v(t) \, dt = \int_1^2 (6t - 6) \, dt
$$

To solve this integral:

$$
D_2 = \left[ 3t^2 - 6t \right]_1^2
$$

Evaluating at the bounds:

$$
D_2 = \left[ 3(2)^2 - 6(2) \right] - \left[ 3(1)^2 - 6(1) \right]
$$

$$
D_2 = \left[ 12 - 12 \right] - \left[ 3 - 6 \right]
$$

$$
D_2 = (0) - (-3) = 3
$$

---

### **Step 5: Calculate the total distance**

Now we add the two parts of the distance:

$$
D = D_1 + D_2 = 3 + 3 = 6
$$

Thus, the total distance traveled by the particle between $ t = 0 $ and $ t = 2 $ is:

$$
\boxed{6}
$$

---

### **Conclusion:**

The total distance traveled by the particle is $ 6 $ units.
