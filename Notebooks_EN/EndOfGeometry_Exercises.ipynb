\textbf{15. Equations of Second-Order Curves (Conic Sections)}
$$

\textbf{Task 1}  
Find the equation of a circle with center at point \( A(1, 2) \) and radius \( r = 3 \).  
The general equation of a circle is:
$$
(x - h)^2 + (y - k)^2 = r^2
$$
where \( (h, k) \) is the center and \( r \) is the radius. Substituting \( h = 1 \), \( k = 2 \), and \( r = 3 \):
$$
(x - 1)^2 + (y - 2)^2 = 9
$$

---

\textbf{Task 2}  
Find the equation of a parabola intersecting the \( OX \)-axis at points \( x = 2 \), \( x = 4 \), and passing through the point \( y(3) = 1 \).  
The general equation of a parabola is \( y = a(x - p)(x - q) \), where \( p \) and \( q \) are the roots. Substituting \( p = 2 \) and \( q = 4 \), and using the point \( (3, 1) \) to find \( a \):
$$
y = a(x - 2)(x - 4)
$$
Substitute \( x = 3 \) and \( y = 1 \) into the equation:
$$
1 = a(3 - 2)(3 - 4)
$$
Solving for \( a \), we find \( a = -1 \), so the equation is:
$$
y = -(x - 2)(x - 4)
$$

---

\textbf{Task 3}  
Find the center of the ellipse with the equation:
$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$
First, we complete the square for \( x \) and \( y \). For \( x^2 - 4x \), add and subtract \( 4 \), and for \( 4y^2 - 16y \), add and subtract \( 16 \):
$$
(x^2 - 4x + 4) + 4(y^2 - 4y + 4) = 4
$$
$$
(x - 2)^2 + 4(y - 2)^2 = 4
$$
Thus, the center of the ellipse is \( (2, 2) \).

---

\textbf{Task 4}  
Find the slope (\( m > 0 \)) of the line \( y = mx - 5 \) that is tangent to the circle with the equation \( x^2 + y^2 = 1 \).  
The distance from the center of the circle \( (0, 0) \) to the line must equal the radius of the circle, which is \( 1 \). The equation of the line is \( y = mx - 5 \), so the distance from the origin to the line is:
$$
\text{Distance} = \frac{|0 \cdot m - 0 - 5|}{\sqrt{m^2 + 1}} = \frac{5}{\sqrt{m^2 + 1}}
$$
Equating this distance to the radius \( 1 \):
$$
\frac{5}{\sqrt{m^2 + 1}} = 1
$$
Solving for \( m \), we find \( m = 2 \).

---

\textbf{Task 5}  
Find the intersection points of the hyperbola \( x^2 - y^2 = 1 \) with the ellipse's line \( x^2 + 4y^2 = 6 \).  
We solve the system of equations:
1. \( x^2 - y^2 = 1 \)
2. \( x^2 + 4y^2 = 6 \)

Substitute \( x^2 = y^2 + 1 \) from the first equation into the second equation:
$$
(y^2 + 1) + 4y^2 = 6
$$
Simplifying:
$$
5y^2 + 1 = 6 \quad \Rightarrow \quad 5y^2 = 5 \quad \Rightarrow \quad y^2 = 1
$$
Thus, \( y = \pm 1 \). Substituting \( y = 1 \) and \( y = -1 \) into \( x^2 = y^2 + 1 \), we find \( x^2 = 2 \), so \( x = \pm \sqrt{2} \).

Thus, the intersection points are \( (\sqrt{2}, 1) \), \( (-\sqrt{2}, 1) \), \( (\sqrt{2}, -1) \), and \( (-\sqrt{2}, -1) \).

---

\textbf{Task 6}  
For the given hyperbola \( x^2 - y^2 = 1 \), find the distance between its branches.  
The distance between the branches of a hyperbola is given by \( 2a \), where \( a \) is the semi-major axis. For the equation \( x^2 - y^2 = 1 \), we have \( a^2 = 1 \), so \( a = 1 \). Therefore, the distance between the branches is:
$$
2a = 2
$$

\textbf{16. Equations of Planes in Space}
$$

\textbf{Task 1}  
The plane passes through points \( A(1, 2, 3) \), \( B(3, 4, 5) \), and \( C(2, 1, 4) \). Find the equation of the plane.  
First, we find two vectors in the plane:  
$$
\vec{AB} = B - A = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2)
$$
$$
\vec{AC} = C - A = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1)
$$
Now, find the cross product \( \vec{AB} \times \vec{AC} \) to get the normal vector:
$$
\vec{n} = \vec{AB} \times \vec{AC} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 2 & 2 & 2 \\ 1 & -1 & 1 \end{vmatrix} = \hat{i}(2 - (-2)) - \hat{j}(2 - 2) + \hat{k}(-2 - 2)
$$
$$
\vec{n} = 4\hat{i} + 0\hat{j} - 4\hat{k} = (4, 0, -4)
$$
The equation of the plane is given by:
$$
4(x - 1) + 0(y - 2) - 4(z - 3) = 0
$$
Simplifying:
$$
4x - 4z + 4 = 0 \quad \Rightarrow \quad x - z = -1
$$

---

\textbf{Task 2}  
The plane passes through point \( A(1, 2, 3) \) and is parallel to the plane \( 2x + 3y + 4z = 5 \). Find the equation of the plane.  
Since the planes are parallel, they have the same normal vector. The equation of the plane passing through point \( A(1, 2, 3) \) is:
$$
2(x - 1) + 3(y - 2) + 4(z - 3) = 0
$$
Simplifying:
$$
2x + 3y + 4z = 20
$$

\textbf{Task 3}  
The plane passes through point \( A(1, 2, 3) \) and is perpendicular to the normal vector \( \vec{n} = [2, 3, 4] \). Find the equation of the plane.  
The equation of the plane is given by:
$$
2(x - 1) + 3(y - 2) + 4(z - 3) = 0
$$
Simplifying:
$$
2x + 3y + 4z = 20
$$

\textbf{Task 4}  
We have two planes \( 2x + 3y + 4z = 5 \) and \( 3x + 4y + 2z = 6 \). Find the line of intersection of these planes.  
To find the line of intersection, we solve the system of equations:
1. \( 2x + 3y + 4z = 5 \)
2. \( 3x + 4y + 2z = 6 \)

We eliminate one variable by multiplying the equations appropriately and solving. After simplifying, the parametric equations for the line of intersection are:
$$
x = 1 + t, \quad y = -2 + t, \quad z = 1 - 2t
$$

\textbf{Task 5}  
Write the equation of the plane passing through point \( A(1, 2, 3) \) and parallel to vectors \( \vec{v_1} = [1, 0, 1] \) and \( \vec{v_2} = [0, 1, -1] \).  
The normal vector to the plane is the cross product of \( \vec{v_1} \) and \( \vec{v_2} \):
$$
\vec{n} = \vec{v_1} \times \vec{v_2} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 1 & 0 & 1 \\ 0 & 1 & -1 \end{vmatrix} = \hat{i}(0 - 1) - \hat{j}(1 - 0) + \hat{k}(1 - 0)
$$
$$
\vec{n} = -\hat{i} - \hat{j} + \hat{k} = (-1, -1, 1)
$$
The equation of the plane is:
$$
-1(x - 1) - 1(y - 2) + 1(z - 3) = 0
$$
Simplifying:
$$
-x - y + z = 0
$$

\textbf{Task 7}  
We have the plane \( 2x + 3y + 4z = 5 \) and point \( A(1, 2, 3) \). Find the distance from point \( A \) to this plane.  
The formula for the distance from a point \( (x_1, y_1, z_1) \) to a plane \( Ax + By + Cz + D = 0 \) is:
$$
d = \frac{|Ax_1 + By_1 + Cz_1 + D|}{\sqrt{A^2 + B^2 + C^2}}
$$
For the plane \( 2x + 3y + 4z = 5 \) and point \( A(1, 2, 3) \), we have \( A = 2 \), \( B = 3 \), \( C = 4 \), and \( D = -5 \). Substituting into the formula:
$$
d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}} = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}} = \frac{|15|}{\sqrt{29}} = \frac{15}{\sqrt{29}}
$$

\textbf{Task 8}  
The plane intersects the coordinate axes at points \( A(2, 0, 0) \), \( B(0, 3, 0) \), and \( C(0, 0, 4) \). Find the equation of the plane.

\frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1
$$
Substituting the given points \( A(2, 0, 0) \), \( B(0, 3, 0) \), and \( C(0, 0, 4) \), we find:
$$
\frac{x}{2} + \frac{y}{3} + \frac{z}{4} = 1
$$

\textbf{Task 9}  
Calculate the angle between the plane \( x + y + z = 1 \) and the plane \( x = 0 \) (i.e., the \( yz \)-plane).  
The normal vector to the plane \( x + y + z = 1 \) is \( (1, 1, 1) \), and the normal vector to the \( yz \)-plane is \( (1, 0, 0) \). The angle \( \theta \) between the two planes is given by:
$$
\cos \theta = \frac{\vec{n_1} \cdot \vec{n_2}}{|\vec{n_1}| |\vec{n_2}|}
$$
Substituting the values:
$$
\cos \theta = \frac{(1, 1, 1) \cdot (1, 0, 0)}{\sqrt{1^2 + 1^2 + 1^2} \cdot \sqrt{1^2}} = \frac{1}{\sqrt{3}} \quad \Rightarrow \quad \theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right)
$$

\textbf{Task 1}  
Write the equation of a sphere with center at point \( P(1, 2, 3) \) and radius \( r = 3 \).  
The general equation of a sphere with center \( (h, k, l) \) and radius \( r \) is given by:
$$
(x - h)^2 + (y - k)^2 + (z - l)^2 = r^2
$$
For center \( P(1, 2, 3) \) and radius \( r = 3 \), the equation becomes:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

---

\textbf{Task 2}  
Do the spheres with equations \( x^2 + y^2 + z^2 = 1 \) and \( x^2 + y^2 + z^2 = 2 \) have any common points?  
To find the intersection, we subtract the second equation from the first:
$$
(x^2 + y^2 + z^2) - (x^2 + y^2 + z^2) = 1 - 2
$$
This simplifies to:
$$
0 = -1
$$
This is a contradiction, so the spheres do not have any common points.

---

\textbf{Task 3}  
What curve in space is formed by the intersection of the sphere \( x^2 + y^2 + z^2 = 1 \) with the sphere \( (x - 1)^2 + y^2 + z^2 = 1 \)? Find the equation of this curve.  
The first equation of the sphere is:
$$
x^2 + y^2 + z^2 = 1
$$
The second equation of the sphere is:
$$
(x - 1)^2 + y^2 + z^2 = 1
$$
Expanding the second equation:
$$
(x^2 - 2x + 1) + y^2 + z^2 = 1
$$
Now, subtract the first equation from the expanded second equation:
$$
(x^2 - 2x + 1 + y^2 + z^2) - (x^2 + y^2 + z^2) = 1 - 1
$$
Simplifying:
$$
-2x + 1 = 0 \quad \Rightarrow \quad x = \frac{1}{2}
$$
Thus, the curve of intersection is a circle with \( x = \frac{1}{2} \). Substituting \( x = \frac{1}{2} \) into the equation of the first sphere:
$$
\left(\frac{1}{2}\right)^2 + y^2 + z^2 = 1 \quad \Rightarrow \quad \frac{1}{4} + y^2 + z^2 = 1
$$
Simplifying:
$$
y^2 + z^2 = \frac{3}{4}
$$
Thus, the equation of the curve is:
$$
x = \frac{1}{2}, \quad y^2 + z^2 = \frac{3}{4}
$$

\textbf{Task 4}  
Write the equation of the tangent plane to the paraboloid \( z = (x - 1)^2 + y^2 + 1 \) at point \( P(1, 0, 1) \).  
To find the equation of the tangent plane, we first compute the partial derivatives of \( z \) with respect to \( x \) and \( y \).  
The surface is \( z = (x - 1)^2 + y^2 + 1 \).  
The partial derivatives are:
$$
\frac{\partial z}{\partial x} = 2(x - 1), \quad \frac{\partial z}{\partial y} = 2y
$$
At the point \( P(1, 0, 1) \), we have:
$$
\frac{\partial z}{\partial x} = 2(1 - 1) = 0, \quad \frac{\partial z}{\partial y} = 2(0) = 0
$$
So the equation of the tangent plane at \( P(1, 0, 1) \) is:
$$
z - 1 = 0(x - 1) + 0(y - 0)
$$
Simplifying:
$$
z = 1
$$
Thus, the equation of the tangent plane is \( z = 1 \).
