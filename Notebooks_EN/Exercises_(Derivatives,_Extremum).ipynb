\textbf{Task 2}  
Prove:
$$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \cot(x)
$$

We use the chain rule:
$$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \frac{1}{\sin(x)} \cdot \cos(x) = \cot(x)
$$

\textbf{Task 4}  
Using de l'Hospital's Rule, find the improper limits:

1. For \( \lim_{{x \to 0}} \frac{\sin(x)}{x} \),  
Since \( \frac{0}{0} \) form is indeterminate, apply L'Hopital's Rule:
$$
\lim_{{x \to 0}} \frac{\sin(x)}{x} = \lim_{{x \to 0}} \frac{\cos(x)}{1} = 1
$$

2. For \( \lim_{{x \to \infty}} \frac{\ln(x)}{x} \),  
Apply L'Hopital's Rule:
$$
\lim_{{x \to \infty}} \frac{\ln(x)}{x} = \lim_{{x \to \infty}} \frac{\frac{1}{x}}{1} = 0
$$

3. For \( \lim_{{x \to \infty}} \frac{e^x}{x} \),  
Apply L'Hopital's Rule:
$$
\lim_{{x \to \infty}} \frac{e^x}{x} = \lim_{{x \to \infty}} \frac{e^x}{1} = \infty
$$

\textbf{Task 5}  
In physics, the position of a particle is given by:
$$
x(t) = 3t^2 - 6t + 1
$$
Find the velocity \( V(t) = x'(t) \) and acceleration \( a(t) = V'(t) = x''(t) \) of the particle at time \( t = 2 \).

First, compute the velocity:
$$
V(t) = \frac{d}{dt} \left( 3t^2 - 6t + 1 \right) = 6t - 6
$$

Now, compute the acceleration:
$$
a(t) = \frac{d}{dt} \left( 6t - 6 \right) = 6
$$

At \( t = 2 \):
$$
V(2) = 6(2) - 6 = 6
$$
$$
a(2) = 6
$$

\textbf{Task 1}  
The profit function is \( P(u) = -2u^2 + 50u - 300 \), where \( u \) is the number of units sold. Find the number of units that maximize profit.

To find the extremum, take the derivative:
$$
P'(u) = -4u + 50
$$
Set \( P'(u) = 0 \):
$$
-4u + 50 = 0 \implies u = 12.5
$$

Thus, the number of units that maximize profit is \( u = 12.5 \).

\textbf{Task 2}  
You have 10 meters of string, and you need to use it to enclose the largest possible rectangular. Find the dimensions of the rectangle.

Let the sides of the rectangle be \( x \) and \( y \), where the perimeter is given by:
$$
2x + 2y = 10 \implies x + y = 5
$$
The area \( A \) of the rectangle is:
$$
A = x \cdot y
$$
Substitute \( y = 5 - x \) into the area formula:
$$
A(x) = x(5 - x) = 5x - x^2
$$
Take the derivative and set it equal to zero:
$$
A'(x) = 5 - 2x \implies 5 - 2x = 0 \implies x = 2.5
$$
Thus, the dimensions of the rectangle are \( x = 2.5 \) and \( y = 2.5 \).

Find the extremum of \( f(x) = x^2 + 3x - 5 \).

Take the derivative:
$$
f'(x) = 2x + 3
$$
Set \( f'(x) = 0 \):
$$
2x + 3 = 0 \implies x = -\frac{3}{2}
