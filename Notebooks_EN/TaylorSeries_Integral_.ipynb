# Problem:  
Elimizde 20 metre uzunluğunda bir ip var ve bu ipi kullanarak en büyük alanlı dikdörtgeni kapatmak istiyoruz. Dikdörtgenin boyutlarını bulmamız gerekiyor.  

Problem aşağıdaki şekilde formüle edilebilir:  

## Verilenler:  
Dikdörtgenin çevresi 20 metre olduğuna göre:  
$$
2x + 2y = 20 \quad \text{(burada \(x\) ve \(y\) dikdörtgenin kenar uzunluklarıdır).}
$$

## Amaç:  
Dikdörtgenin alanını maksimize etmek:  
$$
A = x \cdot y
$$

---

## Çözüm:  

1. **Çevre Denkleminden \(y\)'yi \(x\) Cinsinden İfade Edelim:**  
$$
y = \frac{20 - 2x}{2} = 10 - x
$$

2. **Alan Formülünü \(x\)'e Bağlı Hale Getirelim:**  
$$
A(x) = x \cdot (10 - x) = 10x - x^2
$$

3. **\(A(x)\)'i Maksimize Etmek İçin Türevini Alıp Sıfıra Eşitleyelim:**  
$$
A'(x) = 10 - 2x
$$  
$$
10 - 2x = 0 \implies x = \frac{10}{2} = 5
$$

4. **\(x = 5\) Değerini \(y\)'yi Bulmak İçin Yerine Koyalım:**  
$$
y = 10 - x = 10 - 5 = 5
$$

---

## Sonuç:  
- En büyük alan için dikdörtgenin boyutları:  
$$
x = 5 \, \text{m} \quad \text{ve} \quad y = 5 \, \text{m}.
$$  
- En büyük alan:  
$$
A = 5 \cdot 5 = 25 \, \text{m}^2
$$  


# Problem: Taylor Series and Tangent Lines

## 1. Find the Taylor series and visualize obtained functions in GeoGebra:

### 1.1. Taylor Series of \( f(x) = \cos(x) \) around \( x = 0 \) up to the 4th degree.

The Taylor series of \( \cos(x) \) around \( x = 0 \) is given by:

$$
f(x) = \cos(x) \quad \text{around} \quad x = 0
$$

We use the general formula for the Taylor series expansion around \( x = a \):

$$
f(x) = f(a) + f'(a)(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \frac{f^{(3)}(a)}{3!}(x - a)^3 + \frac{f^{(4)}(a)}{4!}(x - a)^4 + \dots
$$

For \( \cos(x) \), the derivatives at \( x = 0 \) are:

$$
f(x) = \cos(x), \quad f'(x) = -\sin(x), \quad f''(x) = -\cos(x), \quad f^{(3)}(x) = \sin(x), \quad f^{(4)}(x) = \cos(x)
$$

At \( x = 0 \):

$$
f(0) = 1, \quad f'(0) = 0, \quad f''(0) = -1, \quad f^{(3)}(0) = 0, \quad f^{(4)}(0) = 1
$$

So the Taylor series up to the 4th degree is:

$$
f(x) = 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

---

### 1.2. Taylor Series of \( h(x) = \frac{1}{1 - x} \) around \( x = 0 \) up to the 4th degree.

The Taylor series for \( \frac{1}{1 - x} \) is a geometric series:

$$
h(x) = \frac{1}{1 - x} = 1 + x + x^2 + x^3 + x^4 + \dots
$$

The Taylor series up to the 4th degree is:

$$
h(x) = 1 + x + x^2 + x^3 + x^4
$$

---

### 1.3. Taylor Series of \( g(x) = \sin(x) \) around \( x = 0 \) up to the 4th degree.

The Taylor series for \( \sin(x) \) around \( x = 0 \) is:

$$
g(x) = \sin(x)
$$

The derivatives of \( \sin(x) \) at \( x = 0 \) are:

$$
g(x) = \sin(x), \quad g'(x) = \cos(x), \quad g''(x) = -\sin(x), \quad g^{(3)}(x) = -\cos(x), \quad g^{(4)}(x) = \sin(x)
$$

At \( x = 0 \):

$$
g(0) = 0, \quad g'(0) = 1, \quad g''(0) = 0, \quad g^{(3)}(0) = -1, \quad g^{(4)}(0) = 0
$$

Thus, the Taylor series up to the 4th degree is:

$$
g(x) = x - \frac{x^3}{6}
$$

---

## 2. Find a tangent line to the function \( f(x) = g(x) \) at \( x_0 \):

The equation of the tangent line to the function \( f(x) \) at a point \( x_0 \) is:

$$
y = f'(x_0)(x - x_0) + f(x_0)
$$

For the function \( g(x) = \sin(x) \), we find the tangent line at a given point \( x_0 \). The derivative of \( g(x) \) is:

$$
g'(x) = \cos(x)
$$

At \( x_0 \), the tangent line equation becomes:

$$
y = \cos(x_0)(x - x_0) + \sin(x_0)
$$

# Problem: Integrals

## 1. Compute the following integrals:

### 1.1. \(\int 1 \, dx\)

The integral of \( 1 \) with respect to \( x \) is:

$$
\int 1 \, dx = x + C
$$

where \( C \) is the constant of integration.

---

### 1.2. \(\int (x^3 + 2) \, dx\)

The integral of \( x^3 + 2 \) with respect to \( x \) is:

$$
\int (x^3 + 2) \, dx = \frac{x^4}{4} + 2x + C
$$

---

### 1.3. \(\int 2 \sin(x) \, dx\)

The integral of \( 2 \sin(x) \) with respect to \( x \) is:

$$
\int 2 \sin(x) \, dx = -2 \cos(x) + C
$$

---

### 1.4. \(\int x \, dx\)

The integral of \( x \) with respect to \( x \) is:

$$
\int x \, dx = \frac{x^2}{2} + C
$$

---

### 1.5. \(\int (e^x - 5) \, dx\)

The integral of \( e^x - 5 \) with respect to \( x \) is:

$$
\int (e^x - 5) \, dx = e^x - 5x + C
$$

---

### 1.6. \(\int (\sin(x) + \cos(x)) \, dx\)

The integral of \( \sin(x) + \cos(x) \) with respect to \( x \) is:

$$
\int (\sin(x) + \cos(x)) \, dx = -\cos(x) + \sin(x) + C
$$

---

### 1.7. \(\int (3 \sin(x) + 3e^x) \, dx\)

The integral of \( 3 \sin(x) + 3e^x \) with respect to \( x \) is:

$$
\int (3 \sin(x) + 3e^x) \, dx = -3 \cos(x) + 3e^x + C
$$

---

### 1.8. \(\int x e^x \, dx\)

To solve \( \int x e^x \, dx \), we apply integration by parts:

$$
\int x e^x \, dx = x e^x - e^x + C
$$

---

### 1.9. \(\int \sqrt{10x} \, dx\)

The integral of \( \sqrt{10x} \) with respect to \( x \) is:

$$
\int \sqrt{10x} \, dx = \frac{2}{3} \cdot \sqrt{10x}^3 + C = \frac{2}{3} \cdot (10x)^{3/2} + C
$$

---

### 1.10. \(\int \cos(x + 3) \, dx\)

The integral of \( \cos(x + 3) \) with respect to \( x \) is:

$$
\int \cos(x + 3) \, dx = \sin(x + 3) + C
$$

---

### 1.11. \(\int e^x \sin(x) \, dx\)

To solve \( \int e^x \sin(x) \, dx \), we use integration by parts twice, or refer to standard results:

$$
\int e^x \sin(x) \, dx = \frac{1}{2} e^x (\sin(x) - \cos(x)) + C
$$

---

### 1.12. \(\int x \sin(x) \, dx\)

We apply integration by parts for \( \int x \sin(x) \, dx \):

$$
\int x \sin(x) \, dx = -x \cos(x) + \sin(x) + C
$$

---

### 1.13. \(\int \sin(x) \, dx\)

The integral of \( \sin(x) \) with respect to \( x \) is:

$$
\int \sin(x) \, dx = -\cos(x) + C
$$

---

### 1.14. \(\int e^{3x} \, dx\)

The integral of \( e^{3x} \) with respect to \( x \) is:

$$
\int e^{3x} \, dx = \frac{1}{3} e^{3x} + C
$$
